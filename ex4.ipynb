신용카드 사기 거래 탐지 : KNeighborsClassifier

```python
import pandas as pd # csv => DataFrame
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# 데이터로드
data = pd.read_csv('creditcard.csv')
print(data.info())
```

```python
# 레이블 분리 : Class => Target
# Class : 0(정상), 1(사기거래)
x = data.drop('Class', axis=1) # 특성
y = data['Class'] # 출력 데이터(레이블)
```

```python
print(x.head())
print(y.head())
```

```python
# 훈련셋과 테스트 셋으로 분할 => 8:2
x_train, x_test, y_train, y_test = train_test_split(x, 
    y, test_size=0.2, random_state=42)
```

```python
# 데이터 스케일링
# 표준화 : 평균 0, 표준편차 1
scaler = StandardScaler() # 스케일러 생성
x_train_scaled = scaler.fit_transform(x_train) # 훈련 데이터 스케일링
x_test_scaled = scaler.fit_transform(x_test) # 테스트 데이터 스케일링
```

```python
# KNN 객체 생성 => 학습
# n_neighbors = 5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)
```

```python
# 예측
y_pred = knn.predict(x_test_scaled)
```

```python
# 성능평가
acc = accuracy_score(y_test, y_pred)
print('accuracy_score : ', acc)

print(classification_report(y_test, y_pred))
```
