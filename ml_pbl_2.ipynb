```python
# 필요한 라이브러리 호출
import pandas as pd # csv 불러오기
from sklearn.model_selection import train_test_split # 학습 및 테스트 데이터 분리
import numpy as np
from sklearn.metrics import mean_absolute_error # mae계산
from sklearn.tree import DecisionTreeRegressor
```

```python
# 데이터 불러오기
data = pd.read_csv('house.csv')
```

```python
print(data)
```

```python
print(data.info())
```

```python
data.isna().sum()
```

```python
# 결측치 처리
# 1460개의 값중 절반인 730개 이상의 값이 결측치인 컬럼은 drop
data = data.dropna(axis=1, thresh = 730)
print(data.info())
# 결과를 확인하면 총 5개의 컬럼이 drop됨
```

```python
# LotFrontage는 평균값으로 대체
data['LotFrontage'] = data['LotFrontage'].fillna(data['LotFrontage'].mean())
print(data.info())
```

```python
# 결측치 개수 확인
print(data.isnull().sum())
```

```python
# 범주형 데이터 처리
# pd.get_dummies를 이용하여 범주형 데이터를 숫자로 변환
pd.get_dummies(data = data)
print(data)
```
